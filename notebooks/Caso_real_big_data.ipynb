# Caso real de Big Data: Netflix

## ¿Qué datos genera?
- **Tipo de datos**: 
  - Historial de visualización (qué series/películas ves, cuánto tiempo).
  - Datos de interacción (pausas, retrocesos, búsquedas).
  - Información del dispositivo y ubicación.
  - Preferencias de usuario (listas, calificaciones).
- **Volumen**: Terabytes diarios, millones de registros por segundo.
- **Frecuencia**: Datos en tiempo real cada vez que un usuario interactúa con la plataforma.

## ¿Quién los usa?
- **Empresas**: Netflix para mejorar su catálogo y recomendaciones.
- **Gobiernos**: En algunos casos, para estudios de consumo cultural.
- **Investigadores**: Analizan tendencias de entretenimiento y comportamiento digital.
- **Usuarios**: Reciben recomendaciones personalizadas basadas en su historial.

## ¿Qué problema resuelve?
- Mejora la **experiencia del usuario** con recomendaciones precisas.
- Reduce **costes** al optimizar servidores y ancho de banda según patrones de uso.
- Permite **crear contenido original** basado en tendencias (ejemplo: *House of Cards* se produjo tras analizar que los usuarios amaban a Kevin Spacey y thrillers políticos).
- Aporta valor estratégico: fideliza clientes y aumenta el tiempo de permanencia en la plataforma.

## ¿Por qué no bastaría Excel?
- **Volumen**: Excel no puede manejar terabytes de datos ni millones de registros simultáneos.
- **Velocidad**: Los datos llegan en tiempo real; Excel no procesa flujos continuos.
- **Variedad**: Se combinan datos estructurados (tablas) y no estructurados (texto, video, logs).
- **Complejidad**: Se requieren algoritmos de machine learning y sistemas distribuidos (Hadoop, Spark), imposibles de implementar en una hoja de cálculo.
