# Ejercicio 1: Clasificación de edad 
- Solicite o defina una edad y determine la categoría correspondiente. 
- logica: se uso if, elif y else. para ejecutar los caso

```python
def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad < 18:
        return "Adolescente"
    elif edad < 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 21
categoria = clasificar_edad(edad)
print(f"Una persona de {edad} años es un {categoria}.")
```

# Ejercicio 3: Calculadora básica
- solicite o defina 2 numeros y una operacion matematica

```python
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
operacion = input("ingrese la operacion (+, -, *, /): ")

if operacion == "+":
    resultado = num1 + num2
elif operacion == "-":
    resultado = num1 - num2
elif operacion == "*":
    resultado = num1 * num2
elif operacion == "/":
    if num2 != 0:
        resultado = num1 / num2
    else:
        resultado = "Error: división entre cero"
else:
    resultado = "Operación no válida"

print("El resultado es:", resultado)
```

# Ejercicio 8: Lista de números positivos y negativos
- Defina o solicite 10 números. 

```python
numeros = [5, -3, 0, 7, -1, 9, 0, -6, 4, -2]
positivos = 0
negativos = 0
ceros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        ceros += 1
        
print("Resumen de la lista:")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Ceros: {ceros}")
```

# Ejercicio 9: Mayor y menor de una lista 
- Cree una lista con varios números enteros.
- en el max devuelve el valor mas grande de la lista y el min devuelve el valor minimo ambos en una lista

```python

numeros = [2, 4, -7, 0, 8, 10, -100, 1, 12, 6]

mayor = max(numeros)

menor = min(numeros)

diferencia = mayor - menor

print("Lista de números:", numeros)
print("Número mayor:", mayor)
print("Número menor:", menor)
print("Diferencia entre mayor y menor:", diferencia)
```

# Ejercicio 10: Diccionario de estudiante 
- Cree un diccionario que almacene nombre, edad, carrera, promedio y estado.

Para acceder a un valor, se usa la clave entre corchetes

# Ejercicio 11: Función para calcular descuento 


```python

def calcular_descuento(precio, descuento):
    return precio - (precio * descuento / 100)

print(calcular_descuento(568, 16.5))
```
